```python
import pandas as pd

df = pd.read_csv('20210820_20220228_Anue_wdstock.csv', dtype={
    'publishAt': 'str',
    'title': 'str',
    'content': 'str'
})
df
```

```python
df['publishAt'] = pd.to_datetime(df['publishAt'], unit='s')
df
```

```python
import re

pattern = re.compile('&lt;.+?(&gt;|>)')
```

```python
df['content'] = df['content'].str.replace(pattern, "")
df['content'] = df['content'].str.replace('&amp;nbsp;', "")
df
```

```python
df.to_csv('20210820_20220228_Anue_wdstock.csv', encoding='utf-8-sig', index=False)
```
