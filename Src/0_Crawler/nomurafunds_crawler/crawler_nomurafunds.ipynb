```python
import os 
import logging
from tqdm import tqdm
import pandas as pd 

from time import sleep
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import Select
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
```

```python
def get_page(page):
    try:
        page_now_on = int(driver.find_element(by=By.CLASS_NAME, value='current').text[-1])
    except:
        page_now_on = 1
    for i in range(page-page_now_on):
        driver.find_element(by=By.CLASS_NAME, value='pagination-next').click()
        sleep(5)
```

```python
def get_data(row_data):
    name = driver.find_element(by=By.CLASS_NAME, value='appNomuraSubtitle').text
    print('{}.'.format(name))
    driver.find_elements(by=By.CLASS_NAME, value='tab-column-4')[1].click()
    sleep(5)

    title = driver.find_elements(by=By.CLASS_NAME, value='title')
    info_list = []
    for t in title:
        info_list.append(t.text)

    if '' in info_list:
        info_list.remove('')

    print('儲存檔案.')
    print('-'*10)
    row_data.append(name)
    for i in info_list:
        row_data.append(i)
    data.append(row_data)
    
    return data
```

```python
def crawler_fund(option, page, page_total):
    while page <= page_total:
        print('現在要爬第{}頁.'.format(page))
        
        if page != 1:
            print('下一頁.')
            get_page(page)
        
        if page != page_total:
            fund_numbers = 10
        else:
            element = driver.find_elements(by=By.XPATH, value='/html/body/app-root/app-main/div/app-primary/main/div/div/div/app-intro-list/div[2]/div[1]/table/tbody/tr')
            fund_numbers = len(element)
            print(fund_numbers)
            
        for fund_number in range(fund_numbers):
            row_data = []

            print('-'*10)
            print('{}:'.format(fund_number), end=' ')
            fund_elements = driver.find_elements(by=By.XPATH, value='/html/body/app-root/app-main/div/app-primary/main/div/div/div/app-intro-list/div[2]/div[1]/table/tbody/tr')
            fund_elements[fund_number].click()
            sleep(5)
            
            data = get_data(row_data=row_data)
            
            print('返回上一頁.')
            driver.back()
            sleep(5)
            get_option(option)
            get_page(page)
        
        page+=1
            
    print('本基金全部爬完.')
    df = pd.DataFrame(data)
    df.to_excel('../data/{}.xlsx'.format(option), index=None)
```

```python
def get_option(option):
    select = Select(driver.find_element(by=By.CLASS_NAME, value='form-control.ng-untouched.ng-pristine.ng-valid'))
    select.select_by_visible_text(option)
    sleep(2)
    driver.find_element(by=By.XPATH, value='/html/body/app-root/app-main/div/app-primary/main/div/div/div/app-intro-list/div[1]/form/div[5]/button').click()
    sleep(5)
```

```python
url = 'https://www.nomurafunds.com.tw/Web/Content/#/primary/fund/fund-list'

driver = webdriver.Chrome(ChromeDriverManager().install())
driver.maximize_window()
driver.get(url)
sleep(5)

select = Select(driver.find_element(by=By.CLASS_NAME, value='form-control.ng-untouched.ng-pristine.ng-valid'))
options = []
for i in select.options:
    options.append(i.text)
options.remove('請選擇基金公司')
options.remove('野村投信')
options.remove('NN (L)')
options.remove('晉達')
options.remove('野村基金(愛爾蘭系列)')
options.remove('荷寶')
options.remove('駿利亨德森')
options.remove('安本標準')
options.remove('鋒裕匯理')

for option in options:
    data = []
    
    print('選擇基金公司：{}.'.format(option))
    get_option(option)
    
    page_total = len(driver.find_elements(by=By.XPATH, value='/html/body/app-root/app-main/div/app-primary/main/div/div/div/app-intro-list/div[2]/div[3]/app-pagination/div/pagination-controls/pagination-template/ul/li'))-2
    if page_total == -2:
        page_total = 1
    
    crawler_fund(option, page=1, page_total=page_total)
    
```
