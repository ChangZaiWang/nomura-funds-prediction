```python
#原檔
import pandas as pd


highest_df = pd.read_csv("超高資產_總交易檔.csv")
highest_datas=highest_df.to_dict('records')

high_df = pd.read_csv("高資產_總交易檔.csv")
high_datas=high_df.to_dict('records')


middle_df = pd.read_csv("中資產_總交易檔.csv")
middle_datas=middle_df.to_dict('records')

low_df = pd.read_csv("低資產_總交易檔.csv")
low_datas=low_df.to_dict('records')
```

```python
#指標
import csv

vix={}
taiex={}
SP500={}

with open('c.VIX.csv', 'r', newline='', encoding='utf-8-sig') as csvfile:
  datas = csv.DictReader(csvfile) 
  for i in datas:
    vix[i['日期']]=i['三日平均']

with open('c.TAIEX.csv', 'r', newline='', encoding='utf-8-sig') as csvfile:
  datas = csv.DictReader(csvfile) 
  for i in datas:
    taiex[i['日期']]=i['三日平均']

with open('c.S&P 500.csv', 'r', newline='', encoding='utf-8-sig') as csvfile:
  datas = csv.DictReader(csvfile) 
  for i in datas:
    SP500[i['Date']]=i['三日平均']
```
